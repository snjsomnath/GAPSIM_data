# Visualizing different distributions used to procedurally generate activity sequences

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Parameters
x = np.linspace(-10, 20, 1000)

# Normal Distribution
normal = stats.norm.pdf(x, loc=0, scale=1)

# Bimodal Distribution (sum of two Normal Distributions)
bimodal = stats.norm.pdf(x, loc=-2, scale=1) + stats.norm.pdf(x, loc=3, scale=1)

# Trimodal Distribution (sum of three Normal Distributions)
trimodal = stats.norm.pdf(x, loc=-4, scale=1) + stats.norm.pdf(x, loc=0, scale=1) + stats.norm.pdf(x, loc=4, scale=1)

# Chi-Squared Distribution
chi_squared = stats.chi2.pdf(x, df=2)

# Gamma Distribution
gamma = stats.gamma.pdf(x, a=2)

# Inverse Gaussian Distribution
inv_gaussian = stats.invgauss.pdf(x, mu=1, loc=0, scale=1)

# Weibull Distribution
weibull = stats.weibull_min.pdf(x, c=1.5)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x, normal, label='Normal Distribution', linewidth=1)
plt.plot(x, bimodal, label='Bimodal Distribution', linewidth=1, linestyle='--')
plt.plot(x, trimodal, label='Trimodal Distribution', linewidth=1, linestyle='-.')
plt.plot(x, chi_squared, label='Chi-Squared Distribution', linewidth=1, linestyle=':')
plt.plot(x, gamma, label='Gamma Distribution', linewidth=1, color='orange')
plt.plot(x, inv_gaussian, label='Inverse Gaussian Distribution', linewidth=1, color='green')
plt.plot(x, weibull, label='Weibull Distribution', linewidth=1, color='red')
plt.title('Various Probability Distributions', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(-10, 20)
plt.ylim(0, 1.2)
plt.tight_layout()
# Remove the frame
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns

# Set the style of seaborn
sns.set(style="whitegrid")

# Parameters
x = np.linspace(-10, 20, 1000)

# Create data for distributions
data = {
    'Normal Distribution': stats.norm.pdf(x, loc=0, scale=1),
    'Bimodal Distribution': stats.norm.pdf(x, loc=-2, scale=1) + stats.norm.pdf(x, loc=3, scale=1),
    'Trimodal Distribution': stats.norm.pdf(x, loc=-4, scale=1) + stats.norm.pdf(x, loc=0, scale=1) + stats.norm.pdf(x, loc=4, scale=1),
    'Chi-Squared Distribution': stats.chi2.pdf(x, df=2),
    'Gamma Distribution': stats.gamma.pdf(x, a=2),
    'Inverse Gaussian Distribution': stats.invgauss.pdf(x, mu=1, loc=0, scale=1),
    'Weibull Distribution': stats.weibull_min.pdf(x, c=1.5)
}

# Plotting using seaborn
plt.figure(figsize=(10, 5))
for key, value in data.items():
    sns.lineplot(x=x, y=value, label=key)
plt.title('Various Probability Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
